```python
import numpy as np
```

```python
def checkerboard(n):
  c = np.ones((n,n))
  for i in range(n):
    for j in range(n):
      if (i+j)%2 == 1:
        c[i,j] = 0
  return c

print(checkerboard(10))
```

```python
array1 = np.array([0, 10, 20, 40, 60])
array2 = [10, 30, 40]
print(np.intersect1d(array1, array2))
```

```python
x = np.array([10, 10, 20, 20, 30, 30])
print(x)
print(np.unique(x))
x = np.array([[1, 1], [2, 3]])
print("\n",x)
print(np.unique(x))
```

```python
array1 = np.array([0, 10, 20, 40, 60, 80])
print(array1)
array2 = [10, 30, 40, 50, 70, 90]
print(array2)
print(np.setdiff1d(array1, array2))
```

```python
array1 = np.array([0, 10, 20, 40, 60, 80])
print(array1)
array2 = [10, 30, 40, 50, 70]
print(array2)
print(np.setxor1d(array1, array2))
```
