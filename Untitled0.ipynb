```python
class Vehicle:
  def __init__(self, name, max_speed, mileage, capacity):
    self.name = name
    self.max_speed = max_speed
    self.mileage = mileage
    self.capacity = capacity
    self.color = "white"
  def seating_capacity(self, capacity):
    return f'The seating capacity of a {self.name} is {capacity} passanger'
  def fare(self):
    return self.capacity *100
```

```python
class Bus(Vehicle):
  def __init__(self, name, max_speed, mileage, capacity):
    super().__init__(name, max_speed, mileage, capacity)
  def __str__(self):
    return("Vehicle name: {name}  Speed: {max_speed}  Mileage: {mileage}".format(name=self.name, max_speed=self.max_speed, mileage=self.mileage))

  def fare(self):
    return self.capacity *110

b1 = Bus("Volvo", 180, 5000, 50)
print(b1)

print(b1.seating_capacity(50))
print(b1.fare())
print(type(b1))
issubclass(Bus, Vehicle)
```

```python
class Car(Vehicle):
  def __init__(self, name, max_speed, mileage):
    super().__init__(name, max_speed, mileage)
  def __str__(self):
    return("Vehicle name: {name}  Speed: {max_speed}  Mileage: {mileage}".format(name=self.name, max_speed=self.max_speed, mileage=self.mileage))

b1 = Bus("Audi Q5", "120", "25000")
print(b1)

print(b1.seating_capacity(5))
```
