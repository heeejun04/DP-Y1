```python
#6
def charfreq(str):
  chardict = dict()
  for char in str:
    if not char in chardict.keys():
      chardict(char) = 1
    else:
      chardict(char) += 1
  return chardict
charfreq("google.com")
```

```python
#10
def checkerboard(n):
  c = np.ones((n,n))
  for i in range(n):
    for j in range(n):
      if (i+j)%2 == 1:
        c[i,j] = 0
  return c

print(checkerboard(10))
```
