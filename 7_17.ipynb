```python
def ls(s):
  sl = []
  for i in range(len(s)):
    cl = []
    for j in range(i, len(s)-i):
      if not s[j] in cl:
        cl.append(s[j])
      else:
        break
    sl.append(len(cl))
  return max(sl)

ls("assasasab")
```

```python
def maxWater(l):
  water = []
  for i in range(len(l)):
    for j in range(i+1, len(l)):
      water.append( (j-i) * min([l[i], l[j]]))
  return max(water)

maxWater([1,8,6,2,5,4,8,3,7])
```

```python
def nextPermutation(l):
  n = len(l) - 2
  while n >= 0 and l[n+1] <= l[n]:
    n = n - 1
  if n >= 0:
    m = len(l) - 1
    while l[m] <= l[n]:
      m = m - 1
    l[n], l[m] = l[m], l[n]
  i = n+1
  j = len(l) - 1
  while i < j:
    l[i], l[j] = l[j], l[i]
    i = i + 1
    j = j - 1
  return l

nextPermutation([1,2,3,4,5])
```

```python
a = "-42"
def num(a): 
  num = ""
  for c in a:
    if c == "-":
      num = num + c
    if c.isdigit():
      num = num + c
  print(num) 

num(a)
```

```python
num = lambda a: int("".join([c for c in a if c == "-" or c.isdijit()]))
num("-41 in minus forty two")
```

```python
def product(num1,num2):
  num1 = num(num1)
  num2 = num(num2)
  return num1 * num2

product("12","24")
```

```python
def largestSubarray(l):
  max = 0
  for i in range(len(l)):
    for j in range(i+1, len(l)):
      if sum(l[i:j+1]) > max:
        max = sum(l[i:j+1])
  return max
```

```python
def generate(numRows):
        result=[]
        for i in range(numRows):
            if(i ==0):
                row = [1]
                result.append(row)
            else:
                row1 = [1]
                j = 1
                while(j < i):
                    row1.append(row[j-1] + row[j])
                    j+=1
                row1.append(1)
                result.append(row1)
                row = row1
        return result

generate(5)
```
